## Loading the dataset

```python
import pandas as pd
url = 'https://drive.google.com/uc?id=1bXz_ksbuLRFP9wDZBE53MyeQi2Ko54PI'
dfs = pd.read_csv(url)
dfs.head()
```

### Dropping all rows with less than or equal to three percentage of missing values in their columns

```python
dfs.dropna(subset=['color', 'director_name', 'num_critic_for_reviews', 'duration',
       'director_facebook_likes', 'actor_3_facebook_likes', 'actor_2_name',
       'actor_1_facebook_likes', 'genres', 'actor_1_name',
       'movie_title', 'num_voted_users', 'cast_total_facebook_likes',
       'actor_3_name', 'facenumber_in_poster', 'plot_keywords',
       'movie_imdb_link', 'num_user_for_reviews', 'language', 'country',
       'title_year', 'actor_2_facebook_likes',
       'imdb_score',  'movie_facebook_likes'], inplace=True)
```

```python
pd.DataFrame({'dtypes':dfs.dtypes,'null_count':dfs.isnull().sum(),'unique_values':dfs.nunique(),
             'percent_of_null_values':dfs.isnull().mean()*100})
```

## Imputing the missing values in columns 'gross','budget','content_rating,'aspect_ratio' which has more than three percentage of missing values in its columns

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.displot(dfs['gross'],kde=True)
plt.show()
print('gross_mean_value',dfs['gross'].mean())
print('gross_median_value',dfs['gross'].median())
```

```python
dfs['gross'].fillna(dfs['gross'].median(),inplace=True)
```

```python

sns.displot(dfs['budget'],kde=True)
plt.show()
print('budget_mean_value',dfs['budget'].mean())
print('budget_median_value',dfs['budget'].median())
```

```python
dfs['budget'].fillna(dfs['budget'].median(),inplace=True)
```

```python
dfs['content_rating'].unique()
```

```python
data=list(dfs['content_rating'])
keys=list(dfs['content_rating'].unique())
# plt.pie(data,labels=keys, autopct='%.0f%%')
print(data)
print(keys)
```

```python
x=[]
for i in keys:
  x.append(data.count(i))

plt.pie(x,labels=keys, autopct='%.0f%%')
plt.show()
```

```python
dfs['content_rating'].value_counts().plot(kind='bar')
```

```python
cr=dfs.loc[dfs['content_rating'].isnull()]
cr
```

```python
dfs.loc[dfs['content_rating'].isnull()]['genres'].unique()
```

```python
ai=list(dfs.loc[dfs['content_rating'].isnull()]['genres'].index)
ac=[]
ac1=[]
p=list(cr['genres'])
for k in range(len(p)):
    if ('Crime' in p[k]) or  ('Thriller' in p[k]) or ('Horror' in p[k]) or  ('Sci-Fi' in p[k]):

      ac.append((ai[k],'R'))
      ac1.append('1')
    else:

      ac.append((ai[k],'PG-13'))
      ac1.append('2')
print(len(ac))
ac[:5]
```

```python
for h in ac:
  dfs.loc[h[0],"content_rating"]=h[1]
```

```python
dfs['aspect_ratio']=dfs['aspect_ratio'].astype(str)
sns.displot(dfs['aspect_ratio'],kde=True)
plt.xticks(rotation=90)
plt.show()
print()
print('aspect_ratio_mode_value',dfs['aspect_ratio'].mode())
```
